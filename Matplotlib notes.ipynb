```python
#Used to show multiple chart in one figure
#data1
categories = ['Mon', 'Tue','Wed','Thrus']
sales =[10,20,40,50]
#Data2
y1=[10,20,40,50]
y2=[35,45,65 ,75]
plt.figure(figsize=(5,6))
#first plot -bar chart

plt.subplot (1,2,1) #row #column  #position
plt.scatter(categories,sales)
plt.title("daily sales")
plt.xlabel("Week Days")
plt.ylabel("Sales")
#second chart

plt.subplot(1,2,2)
plt.bar(y1,y2)
plt.title("User Example")
plt.xlabel("User1")
plt.ylabel("User2")
plt.show()


```
