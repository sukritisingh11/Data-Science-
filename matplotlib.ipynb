```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    "Student_ID": [101,102,103,104,105,106,107,108,109,110,
                   111,112,113,114,115,116,117,118,119,120,
                   121,122,123,124,125],

    "Name": ["Amit","Priya","Rahul","Sneha","Karan",
             "Neha","Rohit","Pooja","Vikas","Anjali",
             "Arjun","Meera","Sanjay","Riya","Deepak",
             "Komal","Nitin","Simran","Manish","Kavita",
             "Yash","Isha","Varun","Aarti","Mohit"],

    "Age": [20,21,22,20,23,
            21,22,20,24,23,
            21,22,20,23,24,
            21,22,20,23,24,
            21,22,23,20,21],

    "Course": ["B.Tech","BCA","MCA","B.Sc","B.Tech",
               "MBA","BCA","MCA","B.Tech","MBA",
               "B.Sc","BCA","MCA","MBA","B.Tech",
               "BCA","B.Sc","MBA","MCA","B.Tech",
               "BCA","MBA","B.Sc","MCA","B.Tech"],

    "Marks": [85,78,92,67,88,
              74,81,90,69,95,
              72,84,91,79,87,
              76,89,83,94,68,
              80,77,86,93,82],

    "City": ["Delhi","Mumbai","Lucknow","Jaipur","Pune",
             "Delhi","Noida","Lucknow","Indore","Bhopal",
             "Kanpur","Agra","Patna","Ranchi","Surat",
             "Chennai","Hyderabad","Kolkata","Nagpur","Delhi",
             "Jaipur","Pune","Noida","Lucknow","Mumbai"]
})

# Display DataFrame
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df["Student_ID"], df["Marks"])
plt.xlabel("Student_ID")
plt.ylabel("Marks")
plt.title("Student_ID VS Marks")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(
    df["Age"],
    df["Marks"],
    marker='o',        # Marker style
    linewidth=2,       # Line width
    linestyle='--',    # Line style
    color='blue',      # Line color
    label='Marks'
)

plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()

plt.show()
```

```python

# Calculate average marks for each course
avg_marks = df.groupby("Course")["Marks"].mean()

# Create the bar chart
plt.figure(figsize=(6, 4))

plt.bar(
    avg_marks.index,
    avg_marks.values,
    color="skyblue",
    edgecolor="black"
)

plt.title("Average Marks by Course")
plt.xlabel("Course")
plt.ylabel("Average Marks")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
df['City'].value_counts().plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Number of Students')
plt.title('Number of Students in Each City')
plt.show()
```

```python
top5 = df.nlargest(5, 'Marks')

plt.figure(figsize=(8,5))
plt.barh(top5['Student_ID'].astype(str), top5['Marks'], color='orange')

plt.title("Top 5 Students by Marks")
plt.xlabel("Marks")
plt.ylabel("Student ID")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Marks'],
         bins=5,
         edgecolor='black',
         alpha=0.7)

plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'],
         bins=6,
         cumulative=True,
         color='orange')

plt.title("Cumulative Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['Marks'])

plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

```python
course_count = df['Course'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(course_count,
        labels=course_count.index,
        autopct='%1.1f%%')

plt.title("Students in Each Course")
plt.show()
```

```python
city_count = df['City'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(city_count,
        labels=city_count.index,
        autopct='%1.1f%%',
        explode=[0.1]*len(city_count),
        shadow=True,
        startangle=90)

plt.title("Students by City")
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.boxplot(df['Marks'])

plt.title("Box Plot of Marks")
plt.ylabel("Marks")
plt.show()
```

```python
courses = df['Course'].unique()

data = [df[df['Course']==c]['Marks'] for c in courses]

plt.figure(figsize=(8,5))
plt.boxplot(data, labels=courses)

plt.title("Marks Comparison by Course")
plt.xlabel("Course")
plt.ylabel("Marks")
plt.show()
```

```python
df_sorted = df.sort_values('Student_ID')

plt.figure(figsize=(8,5))
plt.fill_between(df_sorted['Student_ID'],
                 df_sorted['Marks'],
                 color='skyblue',
                 alpha=0.5)

plt.plot(df_sorted['Student_ID'], df_sorted['Marks'], color='blue')

plt.title("Area Chart of Marks")
plt.xlabel("Student ID")
plt.ylabel("Marks")
plt.show()
```

```python
courses = df['Course'].unique()[:3]

course1 = df[df['Course']==courses[0]].sort_values('Student_ID')
course2 = df[df['Course']==courses[1]].sort_values('Student_ID')
course3 = df[df['Course']==courses[2]].sort_values('Student_ID')

x = range(max(len(course1), len(course2), len(course3)))

y1 = course1['Marks'].reset_index(drop=True).reindex(x, fill_value=0)
y2 = course2['Marks'].reset_index(drop=True).reindex(x, fill_value=0)
y3 = course3['Marks'].reset_index(drop=True).reindex(x, fill_value=0)

plt.figure(figsize=(8,5))
plt.stackplot(x, y1, y2, y3, labels=courses)

plt.legend()
plt.title("Stack Plot of Marks by Course")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
