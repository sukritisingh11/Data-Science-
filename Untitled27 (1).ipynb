```python
import pandas as pd
```

```python
student={
    "Rollno":[101,102,103,104,105],
    "Name":["Amit","Priya","Rahul","Sahaa","Karan"],
    "Age":[20,21,19,22,20],
    "City":["Delhi","Mumbai","Pune","Delhi","Jaipur"],
    "Marks":[85,90,78,88,76]
}
print(student)
```

```python
df=pd.DataFrame(student)
```

```python
df.head()
```

```python
# Filtering Sorting & Updating
```

```python
print(df[df["Marks"]>80])
```

```python
print(df[df["City"]=="Delhi"])
```

```python
print(df[df["Age"]==20])
```

```python
print(df[(df["City"]=="Delhi") & (df["Marks"]>80)])
```

```python
print(df[(df["Marks"] >= 80) & (df["Marks"] <= 90)])
```

```python
print(df.sort_values(by=["Marks"]))
```

```python
print(df.sort_values(by=["Marks"], ascending=False))
```

```python
print(df.sort_values(by=["City","Marks"]))
```

```python
def grade(mark):
    if mark >=80:
        print("A")
    elif mark>=60:
        print("B")
    elif mark>=50:
        print("C")
    else:
        print("Fail")
df["grade"]=df["Marks"].apply(grade)
```

```python
print(df)
```

```python
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 40 else "Fail")
```

```python
print(df)
```

```python
df["Marks"]=df["Marks"]+5
print(df)
```

```python
df.fillna(grade)
```

```python
df.loc[df["Name"] == "Rahul", "City"] = "Lucknow"
```

```python
print(df)
```

```python
df
```

```python
df.rename(columns={"Marks": "Percentage"}, inplace=True)
```
